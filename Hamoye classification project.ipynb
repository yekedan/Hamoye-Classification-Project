```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df
```

```python
df=pd.get_dummies(df)
```

```python
df
```

```python
# Sklearn require the target variable in a separate daaset. so,we will drop our target variable from the training dataset and save it in another dataset.
x = df.drop('Churn_Yes',1)
y = df.Churn_Yes
```

# To know the accuracy of the test dataset using the random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
random=RandomForestClassifier(criterion='gini', max_depth=10, min_samples_split=5, min_samples_leaf=1, n_estimators=200, random_state=10)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_cv, y_train, y_cv = train_test_split(x,y, test_size =0.3)
```

```python
random.fit(x_train, y_train)
```

```python
pred=random.predict(x_cv)
```

```python
pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(pred,y_cv))
print('\n')
print(confusion_matrix(pred,y_cv))
```
